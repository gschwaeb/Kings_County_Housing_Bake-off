```python
# does month sold affect price of house?
df.groupby('month').agg({'price': ['mean', 'median', 'std'], 'log_price':['mean', 'median', 'std']})
```

```python
#anova test 
#Null: u_log_price equal for all months
#Alternative: u_log_price not equal for all month
#a = .0
```

```python
stats.f_oneway(df[df['month'] == 1]['log_price'],
                df[df['month'] == 2]['log_price'], 
                df[df['month'] == 3]['log_price'],
               df[df['month'] == 4]['log_price'],
               df[df['month'] == 5]['log_price'],
               df[df['month'] == 6]['log_price'],
               df[df['month'] == 7]['log_price'],
               df[df['month'] == 8]['log_price'],
               df[df['month'] == 9]['log_price'],
               df[df['month'] == 10]['log_price'],
               df[df['month'] == 11]['log_price'],
               df[df['month'] == 12]['log_price']
                )
```

```python
#With pvalue 8.617083012196604e-08, we can reject the null that u_log_price is equal for all months
```
